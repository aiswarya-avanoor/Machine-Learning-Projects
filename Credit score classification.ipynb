```python
df=df.drop(['Gender'],axis=1)
```

## **Label encoding**

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Education']=lb.fit_transform(df['Education'])
```

```python
df['Marital Status']=lb.fit_transform(df['Marital Status'])
```

```python
df['Home Ownership']=lb.fit_transform(df['Home Ownership'])
```

```python
df.dtypes
```

## **Separating input and output samples**

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x,y
```

# **Training and testing data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# **Normalisation**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# **Model creation and Performance evaluation**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
naive=GaussianNB()
vector=SVC()
lst_models=[knn,naive,vector]
```

```python
for i in lst_models:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("******************************************************************")
  print(confusion_matrix(y_test,y_pred))
  print("******************************************************************")
  print(accuracy_score(y_test,y_pred))
  print("******************************************************************")
  print(classification_report(y_test,y_pred))
```

```python
x=df.iloc[:,3:]
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  data=KMeans(n_clusters=i,init="k-means++",random_state=42)
  data.fit(x)
  wcss.append(data.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('no.of clusters')
plt.ylabel('varience')
plt.title('clusters vs varience')
plt.show()
```

```python
data1=KMeans(n_clusters=5,init="k-means++",random_state=42)
y_data=data1.fit_predict(x)
y_data
```

```python
x['clusters']=y_data
x
```

```python
x1=x.iloc[:,:-1].values
y1=x.iloc[:,-1].values
x1,y1
```

```python
# Training and testing data
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
```

```python
# Normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x1_train)
x1_train=scaler.transform(x1_train)
x1_test=scaler.transform(x1_test)
```

```python
# Model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x1_train,y1_train)
y_pred=knn.predict(x1_test)
y_pred
```

```python
# Performance evaluation
from sklearn.metrics import confusion_matrix,accuracy_score
result=confusion_matrix(y1_test,y_pred)
score=accuracy_score(y1_test,y_pred)
result,score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y1_test,y_pred))
```
